```python
import pandas as pd
from sklearn.model_selection import train_test_split

train_data = pd.read_csv("../datafile/train.csv",index_col="Id")
test_data = pd.read_csv("../datafile/test.csv", index_col="Id")

train_data.dropna(axis=0, subset=["SalePrice"], inplace=True)
y = train_data.SalePrice
train_data.drop(["SalePrice"], axis=1, inplace=True)

numerical_cols = [cname for cname in train_data.columns if train_data[cname].dtype in ["int64","float64"]]
X = train_data[numerical_cols].copy()
X_test = test_data[numerical_cols].copy()

X.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

my_pipeline = Pipeline(steps=[("preprocessor",SimpleImputer()), ("model", RandomForestRegressor(n_estimators=50,random_state=0))])
```

```python
from sklearn.model_selection import cross_val_score

scores = -1 * cross_val_score(my_pipeline, X, y, cv=5, scoring="neg_mean_absolute_error")
print(scores.mean())
```

# **Step 1**
---

get_score라는 함수 만들기  
simpleImputer() 사용  
RandomForestRegressor() 사용 / random_state=0으로 지정  / n_estimators라는 parameter를 받아서 사용
fold 개수는 3개로  



### **답**

```python
def get_score(n_estimators):
    my_pipeline = Pipeline(steps=[("preprocessing",SimpleImputer()), ("model", RandomForestRegressor(n_estimators, random_state=0))])
    scores = -1 * cross_val_score(my_pipeline, X, y, cv=3, scoring="neg_mean_absolute_error")
    return scores.mean()
```

# **Step 2**
---

위에서 만든 함수의 parameter를 50, 100 ... 400 까지달리하여 results 변수에 dictionary형식으로 저장하기



```python
results = {}
num_list = [50,100,150,200,250,300,350,400]
for i in num_list:
    results[i] = get_score(i)

print(results)
```

가장 최적의 n_estimator의 값은 200이다
